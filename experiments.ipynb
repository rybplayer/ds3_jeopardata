```python
# normal code
def ttt():
    b = [' '] * 9
    for t in range(9):
        print(f"{b[0]} | {b[1]} | {b[2]}\n--+---+--\n{b[3]} | {b[4]} | {b[5]}\n--+---+--\n{b[6]} | {b[7]} | {b[8]}")
        p = ['X', 'O'][t % 2]
        i = int(input(f"Player {p}, choose (0-8): "))
        if b[i] == ' ': b[i] = p
        if any(b[a] == b[b_] == b[c] != ' ' for a, b_, c in [(0,1,2), (3,4,5), (6,7,8), (0,3,6), (1,4,7), (2,5,8), (0,4,8), (2,4,6)]):
            print(f"Player {p} wins!"); break
    else:
        print("It's a tie!")
ttt()
```
